<h1>Convert a variable to a different data type</h1>

```python
numeric_input = input("Enter numeric input")
string_input = input("Enter string input")

srting_outcome = str(numeric_input) #Converts numeric_input to string_outcome
integer_outcome = int(string_input) #Converts string_input to integer_outcome
float_outcome = float(string_input) #Converts string_input to integer_outcome
print(type(srting_outcome))
print(type(integer_outcome))
print(type(float_outcome))
```

<h1>Convert character date to Date</h1>

```python
from datetime import datetime

date_time_str = '18/09/19 01:55:19'

date_time_obj = datetime.strptime(date_time_str, '%d/%m/%y %H:%M:%S')


print ("The type of the date is now",  type(date_time_obj))
print ("The date is", date_time_obj)
```

<h1>Create plots (Histogram, Scatter, Box Plot)</h1>

<h2>Plot Histogram</h2>

```python
#Plot Histogram
import matplotlib.pyplot as plt
import pandas as pd
df=pd.read_csv("C:/Local Disk/my_folders/python/datasets/movies.csv")
#Plots in matplotlib reside within a figure object, use plt.figure to create new figure
fig=plt.figure()
#Create one or more subplots using add_subplot, because you can't create blank figure
ax = fig.add_subplot(1,1,1)
#Variable
ax.hist(df['Audience score %'],bins = 5)
#Labels and Tit
plt.title('Audience score %')
#plt.xlabel('Age')
#plt.ylabel('#Employee')
plt.show()
```
